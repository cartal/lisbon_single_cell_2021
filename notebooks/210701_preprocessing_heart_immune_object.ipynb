### Load required packages

```python
import numpy as np
import pandas as pd
import scanpy as sc
```

### Set up working environment

```python
sc.settings.verbosity = 3
sc.logging.print_versions()
sc.settings.set_figure_params(dpi = 180, color_map = 'RdPu', dpi_save = 300, vector_friendly = True, format = 'svg')
```

### Read in object 

- For the purpose of this excercise, we will focus on the batch alignment of the immune cells from the human heart. 
- The cells were captured using a combination of different protocols (single cell nuclei + total single cell + CD45+ enriched single cells)
- Data was collected from fourteen donors in two different research centres. 

```python
heart_immune = sc.read_h5ad('/Volumes/Bf110/ct5/refseq/lisbon/raw_data/hca_heart_immune_raw.h5ad')
heart_immune
```

### Explore covariates

```python
sc.pl.umap(heart_immune, color = ['age_group', 'donor', 'gender', 'cell_source', 'n_counts', 'n_genes', 'percent_mito', 'percent_ribo', 'region', 'scrublet_score', 'version', 'cell_states', 'scNym_confidence', 'scNym'], size = 1, frameon = False, ncols = 4)
```
